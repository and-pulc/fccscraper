```python
import camelot
import pandas as pd
import numpy as np
```

```python
## Read in all table regions
filepath = './stations/knxv-tv.pdf'
p1t = camelot.read_pdf(filepath, 
                     strip_text=' .\n', flavor='stream', pages="1",
                     table_regions=['0,285,600,0'],
                      columns=['35,52,97,136,210,256,308,360,390,422,456,469,490,513,550'], 
                       edge_tol=75)
p2t = camelot.read_pdf(filepath, 
                     strip_text=' .\n', flavor='stream', pages="2-end",
                      table_regions=['0,700,600,0'],
                      columns=['35,52,97,136,210,256,308,360,390,422,456,469,490,513,550'], 
                       edge_tol=75)
```

```python
## Put all scraped tables into one array to iterate through.
pdf = [p1t[0].df]
for t in p2t:
    pdf.append(t.df)
adz = processTables(pdf)
```

```python
def processTables(tables):
    progColName=""
    ## Create the ad day/times df.
    ads = pd.DataFrame(columns=['StartDate', 'EndDate', 'Weekdays', 'Spots/Week', 'Rate', 'Rating', 'index'])
    ## Create the shows DF.
    shows = pd.DataFrame(columns=['index', 'Title'])
    ## Create final adtimes DF.
    adtimes = []
    ## Add second line of inventory code
    def addNameIfDiff(p):
        if(p[progColName] != nlines[progColName][p.name]):
            p[progColName] = nlines[progColName][p.name] + p[progColName]
        return p
    ## Merge program names into ad days.
    def mergeProgs(i):
        p = shows[shows['index']<=i]
        p = p.iloc[len(p)-1]
        return p['Title']
    ## Create individual records for each ad spot.  
    def expandDays(x):
        adStartDate = datetime.strptime(x['StartDate'], '%m/%d/%y')
        for i, day in enumerate(x['Weekdays']):
            if(day!='-'):
                if(day.isdigit()):
                    for j in range(0, int(day)):
                        adtimes.append({
                                    'Date': (adStartDate + timedelta(days=i)),  
                                    'Rate': x['Rate'],
                                    'Committee': committee,
                                    'Station': station,
                                    'Program': x['Program']
                                   })
                else: ## CODE FOR MTW NOTATION
                    adtimes.append({
                                    'Date': (adStartDate + timedelta(days=i)),  
                                    'Rate': x['Rate'],
                                    'Committee': committee,
                                    'Station': station,
                                    'Program': x['Program']
                                   })
    ## Filter out day times rows.                
    def getDays(row):
        #print(row)
        dayCols = ['StartDate', 'EndDate', 'Weekdays', 'Spots/Week', 'Rate', 'Rating']
        for val in row:
            try:
                ind = dayCols.index(val)
                dayCols.remove(val)
            except:
                pass
        if(len(dayCols) == 0):
            if(len(headers)==0):
                headers.append(row.name)
            if(row.name+1 != len(buys)):
                for k in buys.loc[row.name+1, :]:
                    if k == 'Week:':
                        headers.append(row.name+1)
    def isOrderDetailTable(df):
        returnVal = False
        def findHeaderRow(row):
            headers =  { 'Type1': ['Amount', 'StartDate', 'Description', 'Rate', 'Spots'],
                         'Type2': ['Amount', 'Start', 'InventoryCode', 'Rate', 'Spots'] }
            type1=0
            type2=0
            for htype in headers:
                for val in row:
                    try:
                        ind = headers[htype].index(val)
                        headers[htype].remove(val)
                    except:
                        pass
            print(len(headers['Type2']))
            if(len(headers['Type1']) == 0):
                returnVal = 'Description'
            else:
                if(len(headers['Type2']) == 0):
                    returnVal = 'InventoryCode'
        df.apply(findHeaderRow)
        return returnVal
    ## Filter out program names.
    lastIndex = 0
    for t in tables:
        progColName = isOrderDetailTable(t) ## This both checks if it is a valid table and figures out which
                                            ## WideOrbit header type the station uses.
        print(progColName)
        if progColName != False:
            ## File out inventory code lines
            progs = t.copy()
            progs.columns = progs.iloc[0]
            #progs = progs.loc[1:, 'Start':]
            progs = progs[(progs['Amount'] != '') & (progs[progColName] != '')]
            progs = progs.reset_index()
            nlines = progs[(progs['Ch'] == '') & (progs[progColName] != '')]
            nlines = nlines.reset_index()
            ## Merge in second lines if the stations forms has them.
            if (len(nlines) != 0):
                if (len(nlines) == len(progs)):
                    progs = progs.apply(addNameIfDiff, axis=1)
                else:
                    print("Scraping Error")
            ## Make index run continously accross pages.
            progs['index'] = progs['index'] + lastIndex
            lastIndex = progs.loc[len(progs)-1, :]['index'] + 1
            ## Remove duplicate columns and append to the master list.
            progs = progs.loc[:,~progs.columns.duplicated()]
            progs = progs.rename(columns={progColName:'Title'})
            progs = progs[['index', 'Title']]
            shows = shows.append(progs)
    ## Seperate out the days and times each ad will be on.
    lastIndex = 0
    for t in tables:
        progColName = isOrderDetailTable(t) ## This both checks if it is a valid table and figures out which
                                            ## WideOrbit header type the station uses.
        if progColName != False:
            headers = []
            buys = t.dropna(how='all')
            print(len(buys))
            buys.apply(getDays, axis=1)
            days = buys.iloc[headers]
            days = days.replace('', np.nan)
            days = days.dropna(axis='columns')
            days.columns = days.iloc[0]
            days = days[1:]
            days = days.reset_index()
            days['index'] = days['index'] + lastIndex
            lastIndex = days.loc[len(days)-1, :]['index'] + 1
            ads = ads.append(days)
    print(len(ads))
    ## Add in Program Names
    ads['Program'] = ads['index'].apply(mergeProgs)
    ads.apply(expandDays, axis=1)
    ## Expand each ad buy listing to individual spots
    return pd.DataFrame(adtimes)
```
